```python
# notes:
# d/dx cos(x) = -sin(x)

# d/dx -sin(x) = -cos(x)

# d/dx -cos(x) = sin(x)

# d/dx sin(x) = cos(x)
```

```python
import sympy as sym
from IPython.display import display,Math
import numpy as np
import matplotlib.pyplot as plt
import sympy.plotting.plot as symplot
```

```python
q = sym.symbols('q')

print(sym.diff( sym.cos(q) ))
print(sym.diff( sym.sin(q) ))
```

```python
f = sym.cos(q)

for i in range(0,8):
    display(Math('\\frac{d}{dx}%s = %s' \
                 %(sym.latex(f),sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
# testing math and computation funtions
import sympy.plotting.plot as symplot

f = sym.cos(q)



for i in range(0,4):
    
    if i == 0:
        p = symplot(f,label=sym.latex(f),show=False,line_color=(i/5,i/3,i/5))
#         p = symplot(f,show=False,label=sym.latex(f))
    else:
        p.extend( symplot(f,label=sym.latex(f),show=False,line_color=(i/5,i/3,i/5)) )
    f = sym.diff(f)
p.legend = True
p.xlim = [-3,3]
p.show()
```

```python
# prototype
# f_a = cos(x + sin(x)) + a

x = sym.symbols('x')

for a in range(0,4):
    if a == 0:
        f = sym.cos(x +sym.sin(x)) + a
        p = symplot(f,show=False)
        pdf = symplot(sym.diff(f),show=False)
    else:
        f = sym.cos(x +sym.sin(x)) + a
        p.extend( symplot(f,label=sym.latex(f),show=False) )
        pdf.extend( symplot(sym.diff(f),show=False))

# p.xlim = [-5,5]
pdf.xlim = [-5,5]
p.show()
pdf.show()
```

```python
# plotting derivatives of trig functions
x,a = sym.symbols('x,a')

f = sym.cos( x + sym.sin(x)) + a
colors = 'brkm'

# ----the functions----

for ai in range(0,4):
    if ai==0:
        p = symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[0])
    else:
        p.extend( symplot(f.subs(a,ai),show=False,label='a=' + str(ai),line_color=colors[ai]))

p.title = "The functions"
p.legend = True
p.show()

# ----derivatives of the functions----

for ai in range(0,4):
    if ai==0:
        p = symplot(sym.diff(f.subs(a,ai)),show=False,label='a=' + str(ai),line_color=colors[0])
    else:
        p.extend( symplot(sym.diff(f.subs(a,ai)),show=False,label='a=' + str(ai),line_color=colors[ai]))

p.title = "Their derivatives"
p.legend = True
p.show()
```
