```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

notes
f(x) = 0, if x<=0
     =-2x, if x>0
    
f(x) = 0,  -infinity <x<=0
     = -2x,0 <x<-infinity

```python
from sympy.abc import x

piece1 = 0
piece2 = -2*x
piece3 = x**3/10

fx = sym.Piecewise( (piece1,x<0),(piece2,(x>=0) & (x<3)),(piece3,x>=3))

fxx = sym.lambdify(x,fx)
xx = np.linspace(-3,5,1234)

display(fx)
plt.plot(xx,fxx(xx))
plt.show()
```

```python
# prototype.ploting piecewise funtions
from sympy.abc import x

# symbolic piecewise components
piece1 = x**3
piece2 = sym.log(x,2)
fx = sym.Piecewise( (piece1, x<=0),(piece2, x>0) )

# numeric conversion and definition
fxx = sym.lambdify(x,fx)
xx = np.linspace(-2,2,1000)

# plot and display
# plt.xkcd(xx[1])
# plt.xkcd(.5,75)
plt.rcdefaults() # undoes plt.xkcd()
plt.plot(xx,fxx(xx),'k')
display(fx)
```

```python
# plotting piecewise funtions . styling
f1 = x**3
f2 = sym.log(x,2)

fx = sym.Piecewise( (f1,x<=0),(f2, x>0) )

fxx = sym.lambdify(x,fx)

display(Math('f(x) = ' + sym.latex(fx)))

with plt.xkcd():
    plt.plot(xx,fxx(xx),'k')
plt.xlim([-2,2])
plt.ylim([-10,3])

plt.show()
```
