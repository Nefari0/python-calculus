```python
import sympy as sym
from IPython.display import display,Math
```

```python
# computing partial derivatives
from sympy.abc import x,y

f = (x**2) + (x*y**2)

# partial derivative with respect to x
display(Math('\\frac{\\partial f}{\\partial x} =  %s' %sym.latex(sym.diff(f,x)) ))
# partial derivative with respect to y
display(Math('\\frac{\\partial f}{\\partial x} =  %s' %sym.latex(sym.diff(f,y)) ))
```

```python
# prototyping 3d plots for partial derivatives using sym.plotting.plot3d

# function and var definitions
partDerX = sym.diff(f,x)
partDerY = sym.diff(f,y)

p = sym.plotting.plot3d(f,(x,-3,3),(y,-3,3),title='$f(x,y)=%s$' %sym.latex(f),show=False)
p.extend = sym.plotting.plot3d(partDerX,(x,-3,3),(y,-3,3),title='$f(x,y)=%s$' %sym.latex(partDerX))
p.extend = sym.plotting.plot3d(partDerY,(x,-3,3),(y,-3,3),title='$f(x,y)=%s$' %sym.latex(partDerY))
p.show()
```
