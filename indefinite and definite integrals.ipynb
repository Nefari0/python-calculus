```python
integral bx**a dx = ((bx**(a+1))/(a+1)) + c

example:
integral 4x**(2) dx = ((4x**3)/(3)) + c
```

```python
import sympy as sym
from IPython.display import display,Math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# indefinite integral

x = sym.symbols('x')

f = x

print(sym.integrate(f))
p = sym.plotting.plot(f,show=False)
p.xlim = [0,1]
p.ylim = [0,1]
p.show()
```

```python
# definite integral

sym.integrate(f,(x,0,1))
```

```python
f = (x**3) / (x-2)
intf = sym.integrate(f)
p = sym.plotting.plot(f,show=False)
p.extend( sym.plotting.plot(intf,(x,2.1,10),show=False,line_color='r'))

p[0].label = '$f(x) = %s$' %(sym.latex(f))
p[1].label = '$\\int f(x) dx = %s$' %(sym.latex(intf))

p.ylim = [-200,200]
p.legend = True
p.show()
```

```python
# prototype.computing indefinite integrals
x = sym.symbols('x')
fx = 2*x**3 + sym.sin(x) # function
deriv = sym.diff(fx) # derivative
intf = sym.integrate(deriv) # inregral

display(fx)
display(deriv)
display(intf)
```

```python
# proving endefinite integrals

# unit definition
x = sym.symbols('x')

f = 2*x**2 + sym.sin(x) # funtion
df = sym.diff(f) # derivative
idf = sym.integrate(df) # integral

display(Math('f(x) = %s' %sym.latex(f)))
display(Math('f\'= %s' %sym.latex(df)))
display(Math('\\int (f\')dx = %s' %sym.latex(idf)))
```
