```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math
```

```python
x = sym.symbols('x')

fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx,x,lim_pnt)
display(Math('\\lim_{x\\to %g} %s = %g' %(lim_pnt,sym.latex(fx),lim)))
```

```python
fxx = sym.lambdify(x,fx) # lambdify bridges sympy ans numaric modules
xx = np.linspace(-5,5,200)

plt.plot(xx,fxx(xx)) # fxx(xx) inputs xx into function fxx through lambdify
plt.plot(lim_pnt,lim,'ro')
print(lim)
```

```python
# computing limit where plugging value in for x puts 0 in denominator
x = sym.symbols('x')
fx = (x**2)/(x-2)
fxx = sym.lambdify(x,fx)
xx = np.linspace(1,3,102)

lim_pnt = 2
lim = sym.limit(fx,x,lim_pnt,dir='-')

plt.plot(xx,fxx(xx))
display(Math('\\lim_{x\\to %g^-} %s = %g' %(lim_pnt,sym.latex(fx),lim)))
```

```python
# prototype
# unit definition
e,x = sym.symbols('e,x')
lim_pnt = 4


# symbolic function definition
fx = sym.exp(-x) * sym.sqrt(x + 1)
gx = sym.cos(x + sym.sin(x))

# numaric function conversion
fxx = sym.lambdify(x,fx)
gxx = sym.lambdify(x,gx)
xx = np.linspace(0,10,200)

# function plots
plt.plot(xx,fxx(xx))
plt.plot(xx,gxx(xx))

# compute function limits
limfx = sym.limit(fx,x,lim_pnt) # limit of fx
limgx = sym.limit(gx,x,lim_pnt) # limit of gx

lims = sym.limit(fx/gx,x,lim_pnt)
discreet_lims = limfx/limgx

# plot limits
plt.plot(lim_pnt,limfx,'ro')
plt.plot(lim_pnt,limgx,'bo')
plt.plot(lim_pnt,lims,'go')
plt.plot(lim_pnt,discreet_lims,'yo')

display(Math('%g' %(discreet_lims)))
display(Math('\\lim_{x\\to %g} \\frac{%s}{%s} = %g' \
             %(lim_pnt,sym.latex(fx),sym.latex(gx),lims)))
# display(Math('\\lim_{x\\to %g^-} %s = %g' %(lim_pnt,sym.latex(fx),limfx)))
# display(Math('\\lim_{x\\to %g^-} %s = %g' %(lim_pnt,sym.latex(gx),limgx)))
```

```python
# computing limits part 1
x = sym.symbols('x')

fx = sym.sqrt(x+1)*sym.exp(-x)
gx = sym.cos(x + sym.sin(x))

fxx = sym.lambdify(x,fx)
gxx = sym.lambdify(x,gx)

xx = np.linspace(0,10,100)

plt.plot(xx,fxx(xx),label='f(x)')
plt.plot(xx,gxx(xx),label='g(x)')

plt.legend()
plt.show()

```

```python
# computing limits part 2
# compute limits of f and g seperately

lim_pnt = 5

lim_fx = sym.limit(fx,x,lim_pnt)
lim_gx = sym.limit(gx,x,lim_pnt)

# compute limit of f/g
lim_fgx = sym.limit(fx/gx,x,lim_pnt)

display(Math('\\frac{lim_{x \\to %g f(x)}}{lim_{x \\to %g g(x)}} = \\frac{%g}{%g} = %g' \
             %(lim_pnt,lim_pnt,lim_fx,lim_gx,lim_fx/lim_gx)))
display(Math('\\lim_{x\\to %g} \\frac{f(x)}{g(x)} = %g' %(lim_pnt,lim_fgx)))
```
