```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math
```

```python
notes:
rule for computing derivative of polynomial:
d/dx (ax**n) = nax**(n-1) where 'a' is some coeficient
```

```python
x = sym.symbols('x')

fx = x**2
dfx = sym.diff(fx)
```

```python
# leibniz notation
display(Math('f(x) = %s, \\quad \\frac{df}{dx} = %s'\
             %(sym.latex(fx),sym.latex(dfx))))
```

```python
# lagrange notation
display(Math('f(x) = %s, \\quad f\'= %s'\
             %(sym.latex(fx),sym.latex(dfx))))
# often pronounce "f prime"
```

```python
# newton notation
display(Math('f(x) = %s, \\quad \\dot{f}= %s'\
             %(sym.latex(fx),sym.latex(dfx))))
```

```python
# testing code functions
import sympy.plotting.plot as symplot

fx = 3 - x**3

p = symplot(fx,(x,-3,3),show=False)
p.extend( symplot(sym.diff(fx),show=False))
p[1].line_color = 'r'
p[0].label = '$f(x) = %s$' %sym.latex(fx)
p[1].label = '$f(x) = %s$' %sym.latex(sym.diff(fx))

p.legend = True
p.ylim = [-10,10]
p.xlim = [-3,3]
p.show()
```

```python
# prototype
x = sym.symbols('x')
fx1 = 3 + 2*x - 5*x**2 + 7*x**4
gx1 = 4*x**2 + x**5

fxd = sym.diff(fx1)
gxd = sym.diff(gx1)
display(fxd+gxd)
display(sym.diff(fx1+gx1))
display(sym.diff(fx1*gx1))
display((fxd*gx1) + (gxd*fx1))
display(fxd*gxd)
```

```python
# computing derivatives of polynomials
x = sym.symbols('x')

f = 3 + 2*x - 5*x**2 + 7*x**4
g = 4*x**2 + x**5

df = sym.diff(f)
dg = sym.diff(g)

d_fXg = sym.diff(f*g)
df_X_dg = sym.diff(f) * sym.diff(g)

d_fPlus_g = sym.diff(f+g)
```

```python
# part 2 
# without applying product rule
display(Math('\\text{Without applying the product rule.}'))
display(Math('\\quad (f\\times g)\' = %s' %sym.latex(sym.expand(d_fXg))))
display(Math('\\quad f\' \\times g\' = %s' %sym.latex(sym.expand(df_X_dg))))
# applying product rule
display(Math('\\text{With applying the product rule.}'))
display(Math('\\quad (f\\times g)\' = %s' %sym.latex(sym.expand(d_fXg))))
display(Math('\\quad f\' \\times g + f\\times g\' = %s' \
             %sym.latex(sym.expand(df*g + f*dg))))
```

```python
# part 3
# summation rule
display(Math('(f+g)\' = %s' %sym.latex(sym.expand(d_fPlus_g))))
display(Math('(f\'+g\')\' = %s' %sym.latex(sym.expand(df+dg))))
```
